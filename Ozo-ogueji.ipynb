## Exercise 1

```python
import sympy as sp

# Define symbols
theta1, theta2, lambda_ = sp.symbols('theta1 theta2 lambda_')

# Define the cost function and constraint
J = 5*theta1**2 + 3*theta2**2
g = 4*theta1 + 2*theta2 - 12

# Define the Lagrangian
L = J + lambda_ * g

# Calculate partial derivatives
dL_dtheta1 = sp.diff(L, theta1)
dL_dtheta2 = sp.diff(L, theta2)
dL_dlambda = sp.diff(L, lambda_)

# Solve the system of equations
solution = sp.solve([dL_dtheta1, dL_dtheta2, dL_dlambda], (theta1, theta2, lambda_))

# Output the solution
solution
```

## Exercise 2 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function and constraint
def cost_function(theta1, theta2):
    return 5*theta1**2 + 3*theta2**2

def constraint(theta1, theta2):
    return 4*theta1 + 2*theta2 - 12

# Create grid of values
theta1_vals = np.linspace(-5, 10, 400)
theta2_vals = np.linspace(-5, 10, 400)
T1, T2 = np.meshgrid(theta1_vals, theta2_vals)
Z = cost_function(T1, T2)

# Create contour plot
plt.contour(T1, T2, Z, levels=50, cmap='viridis')
plt.xlabel(r'$\theta_1$')
plt.ylabel(r'$\theta_2$')

# Plot the constraint line
theta2_constraint = (12 - 4*theta1_vals) / 2
plt.plot(theta1_vals, theta2_constraint, label=r'Constraint: $4\theta_1 + 2\theta_2 = 12$', color='red')

# Plot the optimal point
plt.scatter([36/17], [30/17], color='red', label='Optimal solution')

plt.legend()
plt.title('Contour Plot of Cost Function with Constraint Line')
plt.show()
```

## Contour Plot Interpretation:
### The plot shows the trade-off between minimizing the cost and satisfying the constraint. The optimal solution (red dot) balances both objectives, giving us the values of ùúÉ1 and ùúÉ2 that minimize the cost function while ensuring that the constraint is met. The contour plot visually demonstrates how the cost function changes with different values of the model parameters, and how the constraint restricts the feasible region for the solution.
